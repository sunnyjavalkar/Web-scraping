```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd
```

```python
pages=[]
prices=[]
titles=[]
pages_to_scrape=5


for i in range(1,pages_to_scrape+1):
    url = ('https://www.dmart.in/dmart-grocery-1').format(i)
    pages.append(url)
print(pages)
```

```python
for item in pages:
    page = requests.get(item)
    soup = bs4(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
 for i in soup.findAll('a'):
        ttl=i.getText()
        titles.append(ttl)
```

```python
for j in soup.findAll('p', class_='product-listing--discounted-price'):
        price=j.getText()
        prices.append(price)
```

```python
print(prices)
```

```python
data={'Title': titles, 'Prices': prices,}
```

```python
print(data)
```

```python
df = pd.DataFrame.from_dict(data, orient='index')
df.transpose()
```
